```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
class Aero:
    def __init__(self):
        self.weights = None
    def fit(self, x, y, epochs, learning_paramter = 0.0001):
        self.weights = np.random.rand(x.shape[1], 1)
        epoch_list = []
        loss_list = []
        for i in range(epochs):
            # print("Training the model for epoch: ", i+1)
            clear_output(wait=True)
            error_diff = y - np.matmul(x, self.weights)
            loss_diff = np.matmul(x.T, error_diff)
            self.weights = self.weights + (learning_paramter)*(loss_diff*1.0/y.shape[0])
            epoch_list.append(i+1)
            loss_list.append(np.sum(abs(error_diff)/y.shape[0]))
            # print(loss_list)
            plt.scatter(x=epoch_list, y=loss_list, marker='.', c='b')
            plt.show()
            
    def predict(self, x):
        pass
```

```python
from sklearn.datasets import make_regression

X_train, Y_train = make_regression(n_samples=1000, n_features=6)
```
